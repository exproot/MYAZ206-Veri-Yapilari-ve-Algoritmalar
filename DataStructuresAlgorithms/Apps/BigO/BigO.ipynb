$$
f(x) = x^2 + x + 1\quad O(x^2)
$$
$$
f2(x) = x^2\quad O(x^2)
$$
$$
f3(x) = x^3\quad O(x^3)
$$
$$
f4(x) = x^2 + 2x^2 + x^2 = Cg(n) = 4x^2\quad O(x^2)
$$

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def f(n):
    x = np.array(range(1, n+1)) # 1 2 3.... n
    y = list([])
    for i in x:
        y.append(i**2 + i + 1)
    return (x, np.array(y))
```

```python
def f2(n):
    x = np.array(range(1, n+1))
    y = list([])
    for i in x:
        y.append(i**2)
    return(x, np.array(y))
```

```python
def f3(n):
    x = np.array(range(1, n+1))
    y = list([])
    for i in x:
        y.append(i**3)
    return(x, np.array(y))
```

```python
def f4(n):
    x = np.array(range(1, n+1))
    y = list([])
    for i in x:
        y.append(4 * (i**2))
    return(x, np.array(y))
```

```python
n = 10
(x,y) = f(n)
(x2,y2) = f2(n)
(x3,y3) = f3(n)
(x4, y4) = f4(n)
```

```python
print(x4)
print(y4)
```

```python
plt.plot(x,y, color="blue", marker= 'o', linestyle="dashed", label="x2 + x + 1")
plt.plot(x2,y2, color="red", marker= '^', linestyle="dashed", label="x2")
plt.plot(x3,y3, color="black", marker= 'x', linestyle="dashed", label="x3")
plt.plot(x4,y4, color="yellow", marker= 'v', linestyle="dashed", label="4x^2")
plt.xlabel('x')
plt.ylabel('y')
plt.legend()

for i in range(3,len(x), 3):
    plt.text(x[i],y4[i], f'{y4[i]}')
for i in range(3,len(x), 3):
    plt.text(x[i],y3[i], f'{y3[i]}')
```
